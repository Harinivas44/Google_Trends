## Google Trends

```python
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
import plotly.graph_objects as go
```

```python
pytrends = TrendReq(timeout=(10,25))
```

#### Keywords can be entered by seperating commas.....

```python
keywords=[i for i in input('Enter the keywords to see the trends:').split(',')]
```

```python
pytrends.build_payload(keywords)
```

```python
trend_df=pytrends.interest_over_time()
trend_df.head()
```

```python
trend_df.reset_index(inplace=True)
```

```python
trend_df.head()
```

```python
fig = go.Figure(data=go.Scatter(x=trend_df['date'].tolist(), y=trend_df[keywords[0]].tolist(), mode='lines'))
fig.update_layout(title='Interest over period of times : {} '.format(keywords[0]),xaxis_title='Time',yaxis_title='No.of.searches')
fig.show()
```

```python
region_df=pytrends.interest_by_region()
region_df.reset_index(inplace=True)
sort_region_df=region_df.sort_values(keywords[0],ascending=False).head(10)
fig=go.Figure(data=go.Bar(x=sort_region_df['geoName'].tolist(), y=sort_region_df[keywords[0]].tolist()))
fig.update_layout(title='Interest by Regions(Top-10) : {} '.format(keywords[0]),xaxis_title='Country',yaxis_title='No.of.searches')
fig.show()
```

```python
queries_top=pytrends.related_queries()[keywords[0]]['top']
fig=go.Figure(data=go.Bar(x=queries_top['value'].tolist(), y=queries_top['query'],orientation='h'))
fig.update_layout(title='Top Queries: {} '.format(keywords[0]),xaxis_title='Freq',yaxis_title='Top Query')
fig.show()
```

```python
queries_rising=pytrends.related_queries()[keywords[0]]['rising']
fig=go.Figure(data=go.Bar(x=queries_rising['value'].tolist(), y=queries_rising['query'],orientation='h'))
fig.update_layout(title='Rising Queries: {} '.format(keywords[0]),xaxis_title='Rising Query',yaxis_title='Rising Query')
fig.show()
```
